A análise t-SNE produz visualizações bidimensionais onde galáxias com características semelhantes aparecem próximas umas das outras. Vamos entender como interpretar esses resultados:
1. Princípios Fundamentais do t-SNE

    Preserva vizinhanças locais: Pontos próximos no espaço original (alto dimensional) tendem a ficar próximos na projeção 2D

    Não preserva distâncias globais: Agrupamentos distantes no gráfico não necessariamente são diferentes

    É estocástico: Diferentes execuções podem produzir layouts ligeiramente diferentes

2. Elementos Chave para Análise
a) Agrupamentos Naturais

Procure por concentrações densas de pontos no gráfico. Cada agrupamento provavelmente representa:

    Galáxias com morfologias similares (elípticas, espirais, irregulares)

    Galáxias com padrões de cor semelhantes

    Galáxias em estágios evolutivos parecidos

b) Padrões de Separação

Observe como os diferentes agrupamentos se relacionam:

    Agrupamentos bem separados indicam classes distintas

    Transições suaves sugerem um continuum de propriedades

c) Efeito da Perplexidade

Compare os gráficos com diferentes valores de perplexidade:

    Perplexidade baixa (5-15): Enfatiza estruturas locais, pode mostrar muitos pequenos agrupamentos

    Perplexidade alta (30-100): Revela estruturas mais globais, com menos mas maiores agrupamentos

3. Guia Passo a Passo para Interpretação

    Identifique os principais agrupamentos no gráfico t-SNE

    Analise as miniaturas (quando disponíveis) para entender o que cada agrupamento representa visualmente

    Compare com o PCA para ver se os padrões são consistentes ou revelam novas relações

    Varie a perplexidade para entender estruturas em diferentes escalas

    Procure por outliers - pontos isolados podem representar galáxias com características únicas

4. Exemplo Prático de Interpretação

Suponha que seu t-SNE mostre:

    Agrupamento A (canto superior esquerdo): Miniaturas mostram galáxias elípticas vermelhas

    Agrupamento B (centro direito): Galáxias espirais azuis com braços bem definidos

    Região C (inferior): Galáxias irregulares com formação estelar intensa

Isso sugere que o t-SNE está capturando:

    A sequência de Hubble (elípticas → espirais)

    A correlação cor-forma (vermelho/elíptica vs azul/espiral)

    Galáxias starburst como uma classe distinta

5. Limitações e Cuidados

    Não use distâncias absolutas entre agrupamentos como medida de dissimilaridade

    Execute múltiplas vezes para verificar consistência dos padrões

    Combine com outras técnicas como PCA ou UMAP para validação

    Considere o tamanho dos agrupamentos - maiores não necessariamente significam mais comuns, podem refletir apenas maior variabilidade interna

A interpretação ideal combina a análise quantitativa (gráficos t-SNE) com a qualitativa (inspeção visual das imagens nos agrupamentos).

```python
import os
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
from sklearn.decomposition import PCA
from sklearn.manifold import TSNE
import seaborn as sns
from matplotlib.offsetbox import OffsetImage, AnnotationBbox

# Configurações
base_dir = './png'
image_size = (128, 128)  # Tamanho para redimensionamento
n_components_pca = 50    # Número de componentes para redução inicial com PCA
perplexity_values = [5, 15, 30, 50, 100]  # Valores de perplexidade para testar no t-SNE

# Função para carregar e processar imagens
def load_images(image_dir, size):
    image_list = []
    image_files = [f for f in os.listdir(image_dir) if f.endswith('.png')]
    
    for image_name in image_files:
        img_path = os.path.join(image_dir, image_name)
        try:
            img = Image.open(img_path).convert('L')  # Escala de cinza
            img = img.resize(size)
            img_array = np.array(img).flatten()
            image_list.append(img_array)
        except Exception as e:
            print(f"Erro ao processar {image_name}: {e}")
    
    if not image_list:
        raise ValueError("Nenhuma imagem foi carregada. Verifique o diretório.")
    
    return np.array(image_list), image_files

# Função para visualização t-SNE com miniaturas
def plot_tsne_with_images(X_tsne, images, image_size, title):
    fig, ax = plt.subplots(figsize=(20, 15))
    ax.scatter(X_tsne[:, 0], X_tsne[:, 1], alpha=0)
    
    for i, (x, y) in enumerate(X_tsne):
        img = images[i].reshape(image_size)
        im = OffsetImage(img, zoom=0.5, cmap='gray')
        ab = AnnotationBbox(im, (x, y), xycoords='data', frameon=False)
        ax.add_artist(ab)
    
    ax.update_datalim(X_tsne)
    ax.autoscale()
    plt.title(title)
    plt.xlabel("t-SNE 1")
    plt.ylabel("t-SNE 2")
    plt.show()

# Carregar imagens
X, image_names = load_images(base_dir, image_size)

# Redução inicial com PCA (opcional mas recomendado para grandes conjuntos de dados)
pca = PCA(n_components=min(n_components_pca, X.shape[0], X.shape[1]))
X_pca = pca.fit_transform(X)

# Executar t-SNE para diferentes valores de perplexidade
for perplexity in perplexity_values:
    print(f"\nExecutando t-SNE com perplexity={perplexity}...")
    
    tsne = TSNE(n_components=2, 
                perplexity=perplexity, 
                n_iter=1000, 
                random_state=42,
                verbose=1)
    
    X_tsne = tsne.fit_transform(X_pca)
    
    # Plotar resultados
    plt.figure(figsize=(12, 8))
    sns.scatterplot(x=X_tsne[:, 0], y=X_tsne[:, 1], alpha=0.7)
    plt.title(f"t-SNE projection (perplexity={perplexity})")
    plt.xlabel("t-SNE 1")
    plt.ylabel("t-SNE 2")
    plt.show()
    
    # Plotar com miniaturas das imagens (para conjuntos pequenos)
    if len(X_tsne) <= 100:  # Só plotar miniaturas se não for muitas imagens
        plot_tsne_with_images(X_tsne, X, image_size, f"t-SNE com miniaturas (perplexity={perplexity})")

# Comparação com PCA (2 componentes)
pca_2d = PCA(n_components=2)
X_pca_2d = pca_2d.fit_transform(X)

plt.figure(figsize=(12, 8))
sns.scatterplot(x=X_pca_2d[:, 0], y=X_pca_2d[:, 1], alpha=0.7)
plt.title("Projeção PCA (2 componentes)")
plt.xlabel("PC1")
plt.ylabel("PC2")
plt.show()
```
